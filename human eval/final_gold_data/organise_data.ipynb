```python
# Get all Excel files from the final_gold_data folder
folder_path = "../final_gold_data"
excel_files = [f for f in os.listdir(folder_path) if f.endswith('.xlsx') and 'annotations_' in f]

# Create an empty list to store all dataframes
all_dfs = []

# Loop through each Excel file
for file in excel_files:
    file_path = os.path.join(folder_path, file)
    
    # Read the Excel file
    df = pd.read_excel(file_path)
    
    columns_to_drop = ['Notes', 'done']
    if columns_to_drop:
        df = df.drop(columns=columns_to_drop)
    
    # Add to the list of dataframes
    all_dfs.append(df)

# Concatenate all dataframes
final_df = pd.concat(all_dfs, ignore_index=True)

final_df
```

```python
final_df = final_df.sort_values(by=['provider_name', 'comment_id'])
final_df
```

```python
# Save the combined data to a CSV file
output_file = "gold_data.csv"
final_df.to_csv(output_file, index=False)
```
